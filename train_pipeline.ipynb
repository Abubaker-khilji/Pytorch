# define the model

```python
class mysimplenn():
  def __init__(self ,x):
    self.weights = pt.rand(x.shape[1],1,dtype=pt.float64,requires_grad=True)
    self.bias = pt.zeros(1,dtype=pt.float64,requires_grad=True)

  def forward(self,x):
    z= pt.matmul(x,self.weights)+self.bias
    y_pred = z.sigmoid()
    return y_pred

  def lossfunc(self,y_pred,y):
    #clamp predcition to avoid log(0)
    epsilon = 1e-7
    y_pred = pt.clamp(y_pred,epsilon,1-epsilon)

    #calculate loss using cross entropy
    loss = (y_train_tensor*pt.log(y_pred))+(1-y_train_tensor)*pt.log(1-y_pred)
    loss = -loss.mean()
    return loss
```

```python
learning_rate = 0.1
epochs = 25
```

# training pipeline

```python
model = mysimplenn(X_train_tensor)
model.weights
```

```python
model.bias
```

```python
#define loop
for epoch in range(epochs):
# forward pass
   y_pred = model.forward(X_train_tensor)
# loss calculate
   loss = model.lossfunc(y_pred,y_train_tensor)
#backward pass
   loss.backward()
#update weights
   with pt.no_grad():
     model.weights -= learning_rate * model.weights.grad
     model.bias -= learning_rate * model.bias.grad
#zero gradeints
   model.weights.grad.zero_()
   model.bias.grad.zero_()

#print loss in each epoch
   print(f"epoch {epoch + 1}, loss{loss.item()}")
```
